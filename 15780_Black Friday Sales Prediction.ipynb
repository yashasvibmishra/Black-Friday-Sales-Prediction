# PREPROCESSING

```python
# scanning to check for any null values present in the dataset (if any)
df.isnull().sum()
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(-1.0).astype("float32")
df['Product_Category_3'] = df['Product_Category_3'].fillna(-1.0).astype("float32")
```

```python
# checking for null values reamaining after filling of the null values present prior to me handling them
df.isnull().sum()
```

```python
# doing a value based encoding using inbuilt python dictionary
gender_dict = {'M':1,'F':0}
df['Gender'] = df['Gender'].apply(lambda x: gender_dict[x])
df.head()
```

```python
# enhancing the handling of coloumn using various encodings
# such as one hot & label encoding

cols = ['Age', 'City_Category', 'Stay_In_Current_City_Years']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python


X = df.drop(columns=['User_ID', 'Product_ID', 'Purchase'])
y = df['Purchase']
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
def train(model, X, y):
    # train-test splitting of the dataset
    x_train, x_test, y_train, y_test = train_test_split(X, y,random_state=42, test_size=0.25)
    model.fit(x_train, y_train)
    
    # the prediction the results thus obtained
    pred = model.predict(x_test)
    
    # here I attemmpt to cross-validate my findings
    cv_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("*******Results*******")
    print("Mean Squared Error:", np.sqrt(mean_squared_error(y_test, pred)))
    print("Cross Validation Score:", np.sqrt(cv_score))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(normalize=True)
train(model, X, y)
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind='bar', title='Linear Regression Coefficients')
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model, X, y)
features = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
features.plot(kind='bar', title='Significance of features in Decision Tree model')
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_jobs=-1)
train(model, X, y)
features = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
features.plot(kind='bar', title='Significance of features Random Forest model')
```

```python
# implement xgboost 

from xgboost import XGBRegressor
model = XGBRegressor()
train(model, X, y)
features = pd.Series(model.feature_importances_, X.columns).sort_values(ascending=False)
features.plot(kind='bar', title='Significance of features XGBoost model')




```
